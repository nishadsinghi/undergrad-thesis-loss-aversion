```python
import pandas as pd
data = pd.read_csv("data.csv")
data.head()
```

```python
# rearrange columns, remove 'run', rename columns to be consistent with other datasets
dataRequiredColumns = data[['Subject', 'Response', 'RT', 'Gain', 'Loss']]
dataRequiredColumns.columns = ['subj_idx', 'response', 'rt', 'gain', 'loss']
dataRequiredColumns.head()
```

```python
# convert RT to seconds, make losses negative
dataConvertedValues = dataRequiredColumns.copy()
dataConvertedValues['rt'] = dataConvertedValues['rt']/1000
dataConvertedValues['loss'] = dataConvertedValues['loss'] * (-1)
dataConvertedValues.head()
```

```python
dataConvertedValues.to_csv("data_preprocessed.csv")
```
